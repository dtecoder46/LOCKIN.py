### Function to check the runtime speed of each function

```python
# Purpose: to check how fast a certain block of code runs
# Parameters: function
# Return: none

def stopwatch(function):
    start = time.time() # 0 seconds
    exec(f'{function}') # calls a function
    end = time.time() # time it takes for the function to run

    print(f"{end-start} seconds") 
```

### Read in the YAML file content

```python
# Purpose: to extract tasks from a YAML file
# Parameters, Return: none

def extract_yml():

    # get file content

    yml_file = open("task.yaml", "r")
    yml_content = yml_file.read()

    return yml_content

stopwatch("extract_yml()")
```
